```python
import pandas as pd
import numpy as np
import datetime as dt
from datetime import date
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\vorob\Downloads\DataAnalystSQL\Python\DataClear_6\applications.csv')
#Додавання нового поля
df['Rating'] = 0
```

```python
df_score = pd.read_csv(r'C:\Users\vorob\Downloads\DataAnalystSQL\Python\DataClear_6\industries.csv')
df_score
```

```python
#1.1.Прибери дублікати applicant_id
df1=df.drop_duplicates(subset='applicant_id')
df1
```

```python
#1.2.В полі 'External Rating' заповни відсутні значення нулями
df1['External Rating'].fillna(0, inplace=True)
df1
```

```python
#1.3.В полі 'Education level' заповни відсутні значення текстом “Середня”
df1['Education level'].fillna('Середня', inplace=True)
df1
```

```python
#Check isnull
a=df1.loc[df1['Education level'].isna()]  
b=df1.loc[df1['External Rating'].isna()]
print(a,b)
```

```python
#2.Додай до цього DataFrame дані з файлу industries.csv, а саме, рейтинги індустрій
df_score = df_score.set_index('Industry')
df2 = df1.join(df_score, on='Industry', how='outer')
print(df2)
```

```python
#3.Розрахуй рейтинг заявки:
#3.1.Якщо вік заявщика між 35 та 55, до рейтингу додається 20 балів
df2.loc[(df2['Age'] >= 35) & (df2['Age']<= 55), 'Rating'] +=20
#3.2.Якщо заявка була подана не у вихідні, до рейтингу додається 20 балів
#Обрізання часу з дати
df2['Applied at']=df2['Applied at'].str.slice(0,10)
#Зміна типу даних на {datetime}
df2['Applied at'] = pd.to_datetime(df2['Applied at'])
#Метод weekday: 5 Saturday, 6 Sunday
df2.loc[(pd.to_datetime(df2['Applied at']).dt.weekday < 5), 'Rating'] +=20
#3.3.Якщо заявщик одружений, до рейтингу додається 20 балів
df2.loc[(df2['Marital status'] == 'Married'), 'Rating'] +=20
#3.4.Якщо заявщик знаходиться в Києві чи області, до рейтингу додається 10 балів
df2.loc[df['Location']=='Київ чи область', 'Rating'] += 20
#3.5.Значення 'Score' з таблиці industries.csv також додається до заявки (і складає від 0 до 20 балів)
df2['Rating']=df2['Rating']+df2['Score']
#3.6.Якщо 'External Rating' більше чи дорівнює 7, до рейтингу додається 20 балів
df2.loc[df2['External Rating']>=7, ['Rating']]+=20
#3.7.Якщо 'External Rating' менше чи дорівнює 2, з рейтингу віднімається 20 балів
df2.loc[df2['External Rating']<=2, ['Rating']]-=20
df2
```

```python
#Рейтинг має бути числом від 0 до 100
#Рейтинг - сума оцінок заявки по 6 критеріям.
#Рейтинг дорівнює нулю, якщо відсутнє значення 'Amount' або якщо 'External Rating' дорівнює нулю.
df2.loc[df2['Rating']<0, 'Rating'] = 0
df2.loc[df2['Rating']>100, 'Rating'] = 100
df2.loc[df2['Amount'].isnull(), 'Rating'] = 0
df2.loc[df2['External Rating'] ==0, 'Rating'] = 0
df2
df2.to_clipboard()
```

```python
#З DataFrame видалено заявки з рейтингом менше або дорівнює 0
df2 = df2.drop(df2[df2['Rating'] ==0].index)
df2
```

```python
#Застосовано групування за тижнем подачі заявки,
#та побудовано графік середнього рейтингу прийнятих заявок для кожного тижня.
plt.plot(df2.groupby(pd.Grouper(key='Applied at', freq='W'))['Rating'].mean(), linewidth=1, label='Rating, Applied at') 
plt.title('AVG rating / Week')
plt.xlabel('Week')
plt.ylabel('Rating')
plt.legend()
plt.show()
```

```python
#інший тип графіку #та побудовано графік середнього рейтингу прийнятих заявок для кожного тижня.
df2.groupby(pd.Grouper(key='Applied at', freq='W'))['Rating'].mean().plot(kind='bar', title='AVG rating / Week', xlabel='Week', ylabel='Rating')
plt.legend()
plt.show()
```
